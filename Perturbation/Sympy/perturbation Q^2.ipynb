```python
from sympy import *
init_printing(use_latex="mathjax")
```

```python
x,k,x0=symbols('x,k,x0', real=True)
D,e,R,Q0,A,t0,t=symbols('D,e,R,Q0,A,t0,t', real=True, positive=True)

L=Function('L')(x,t)
P1=Function('P1')(x,t)
P2=Function('P2')(x,t)
P3=Function('P3')(x,t)
P4=Function('P4')(x,t)
P5=Function('P5')(x,t)
P6=Function('P6')(x,t)
P7=Function('P7')(x,t)

Q=L+e*P1+e**2*P2+e**3*P3+e**4*P4+e**5*P5+e**6*P6+e**7*P7
```

```python
PDE_long = diff(Q,t)-D*diff(Q,x,2)+e*R*Q**2+2*R*Q0*Q
collect(PDE_long.expand(),e)
```

```python
Q=L+e*P1
PDE_short = diff(Q,t)-D*diff(Q,x,2)+e*R*Q**2+2*R*Q0*Q
print(latex(collect(PDE_short.expand(),e)))
```
